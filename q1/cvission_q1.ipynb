```python
q2dict = {}

student_count = int(input("enter student count: "))
lesson_count = int(input("enter lesson count: "))

for student in range(student_count):
  key = input("enter student name: ")
  value = 0;
  for score in range(lesson_count):
    value += float(input(f"enter score {score+1}: "))
  q2dict.update({ key : value / lesson_count})

print("-------------")
for key,value in q2dict.items():
  print(f"{key} grade is {value}")
```

```python
import pandas as pd
data = pd.read_csv("student.csv",names=["name","gender","score"])
gpa = data.groupby(['name'])["score"].mean()
mean_gpa = gpa.mean()
max_gpa = gpa.max()
max_gender = data[data["name"] == gpa.idxmax()]["gender"].values[0]
female_average = data[data['gender'] == "female"]["score"].mean()
print(f"gpa: {gpa}")
print("------------")
print(f"mean gpa: {mean_gpa}")
print("------------")
print(f"max gpa: {max_gpa}")
print("------------")
print(f"max gender: {max_gender}")
print("------------")
print(f"female average: {female_average}")
print("------------")
```

```python
average_gender = data.groupby("gender").agg({"score": ["mean", "std"]})
print(average_gender)
average_gender["score"]["mean"].plot(kind='bar', yerr=average_gender["score"]["std"], capsize=4, color=['pink', 'blue'], alpha=0.7)

plt.title('std and mean')
plt.ylabel('mean')
plt.xlabel('gender')

plt.show()
```
